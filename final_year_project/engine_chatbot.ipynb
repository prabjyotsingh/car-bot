```python
# Import the necessary libraries
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer

# Create a new chatbot instance
chatbot = ChatBot(
    'CarAssistantBot',  # The name of your chatbot
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    logic_adapters=[
        'chatterbot.logic.BestMatch',   # Best match logic adapter
        'chatterbot.logic.MathematicalEvaluation',  # Handle math questions
    ],
    database_uri='sqlite:///database.db'  # Saves the chat history in a database
)

# Create a new trainer for the chatbot
trainer = ChatterBotCorpusTrainer(chatbot)

# Train the chatbot with the English corpus
trainer.train('chatterbot.corpus.english')

# Add custom data for car-related questions
trainer.train(
    "I am a car engine assistant.\n"
    "When should I change my engine oil? You should change your oil every 5,000 km.\n"
    "How to prevent engine overheating? Make sure the coolant levels are adequate.\n"
    "What to do if the car won't start? Check the battery and fuel system.\n"
    "How to improve fuel efficiency? Regular maintenance and proper tire pressure help.\n"
)

print("Chatbot created and trained successfully!")
```
